```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot([0, 1], [0, 3], label = 'line', color = 'green')
plt.xlabel('Oś X')
plt.ylabel('Oś Y')
plt.title('Funkcja liniowa')
_ = plt.show()
```

```python
plt.plot([0, 1, 2], [0, 3, 2], label = 'line', color = 'green')
plt.xlabel('Oś X')
plt.ylabel('Oś Y')
plt.title('Funkcja liniowa')
_ = plt.show()
```

```python
przedzial = np.arange(-3, 3, 0.1)
funkcja_sinus = np.sin(przedzial)
funkcja_cosinus = np.cos(przedzial)
funkcja_tangens = np.tan(przedzial)

plt.plot(przedzial, funkcja_sinus)
plt.plot(przedzial, funkcja_cosinus)
plt.plot(przedzial, funkcja_tangens)
```

```python
plt.bar(x = [0, 1, 2, 3], height = [1, 2, 3, 4])
```

```python
plt.barh(y = [0, 1, 2, 3], width = [1, 2, 3, 4])
```

```python
przedzial = np.linspace(0, 64)
funkcja = np.exp(przedzial)

plt.scatter(x = przedzial, y = funkcja)
```

```python
import seaborn as sns
sns.set()
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
tips_df.info()
tips_df.describe().T
tips_df.describe(include = ['category']).T
```

```python
_ = sns.relplot(data = tips_df, x = 'total_bill', y = 'tip')
```

```python
sns.set(font_scale = 1.2)
```

```python
_ = sns.relplot(data = tips_df, x = 'total_bill', y = 'tip', size = 'size', hue = 'size', col = 'time', row = 'sex')
```

```python
_ = sns.catplot(data = tips_df, x = 'day', kind = 'count')
```

```python
titanic_df = sns.load_dataset('titanic')

titanic_df.head()
```

```python
_ = sns.catplot(data = titanic_df, x = 'deck', kind = 'count', hue = 'deck', palette = 'Blues')
```
