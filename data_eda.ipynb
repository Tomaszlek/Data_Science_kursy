```python
import numpy as np
import pandas as pd
```

```python
# Import data
df = pd.read_csv('ratings_with_features.csv')
```

```python
ratings = pd.DataFrame(df.groupby('book_title')['rating'].mean()).sort_values(ascending=False, by='rating')
ratings['num of ratings'] = pd.DataFrame(df.groupby('book_title')['rating'].count())
ratings.sort_values('num of ratings', ascending=False)[:10000]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x="num of ratings", y="book_title", 
            data=ratings.sort_values(by='num of ratings', ascending=False)[:10].reset_index())
plt.xlabel('Number of ratings')
plt.ylabel('Book Title')
```

```python
# Number of ratings distribution


sns.histplot(ratings, x='num of ratings', bins=50)

plt.title('Number of ratings distribution')
```

```python
# Rating distribution
sns.histplot(ratings, x='rating', bins=30)
plt.title('Rating Distribution')

plt.xlabel('Rating')
```

```python
h = sns.jointplot(x='rating', y='num of ratings', data=ratings)

h.set_axis_labels('Rating', 'Number of Ratings', fontsize=10)
```

```python
ratings[ratings['num of ratings'] > 40].sort_values(by='rating', ascending=False)
```

```python
sns.barplot(x="rating", y="book_title", 
            data=ratings[ratings['num of ratings'] > 40][:10].sort_values(by='rating', ascending=False).reset_index())
plt.xlabel('Rating')
plt.ylabel('Book Title')
plt.title('Top rated books with Number of ratings more than 40')
```
