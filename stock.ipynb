```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web
```

```python
def get_financial_data(company):
  return web.DataReader(name = company, data_source = 'stooq')

apple = get_financial_data('AAPL')
amazon = get_financial_data('AMZN')
facebook = get_financial_data('META')
tesla = get_financial_data('TSLA')
google = get_financial_data('GOOGL')
uber = get_financial_data('UBER')

facebook.describe()
```

# Preprocessing


```python
apple.columns = ['apple_' + col.lower() for col in apple.columns]
amazon.columns = ['amazon_' + col.lower() for col in amazon.columns]
facebook.columns = ['facebook_' + col.lower() for col in facebook.columns]
tesla.columns = ['tesla_' + col.lower() for col in tesla.columns]
google.columns = ['google_' + col.lower() for col in google.columns]
uber.columns = ['uber_' + col.lower() for col in uber.columns]

facebook
```

```python
df = pd.concat(objs = [apple, amazon, google, facebook, tesla, uber], axis = 1)

df.head()

#funkcja dla normalnego formatowania
pd.set_option('display.float_format', lambda x: f'{x:.2f}')
df.head()

df.corr()

uber_6 = uber[uber.index.month == 6]
uber_6

uber_7 = uber[uber.index.month == 7]
uber_7

uber_6_7 = pd.concat(objs = [uber_6, uber_7], axis = 0)  # append nie ma w pandasie nowym

uber_6_7
```
